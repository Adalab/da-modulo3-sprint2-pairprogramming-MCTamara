4. Realiza un análisis de la mortalidad infantil (infant_mortality) y clasifica los países en tres categorías: baja, moderada y alta mortalidad infantil.

```python
bajo = np.percentile(df['infant_mortality'], 33)
alto = np.percentile(df['infant_mortality'], 66)


def mortalidad_inf (datos):
    if datos >= bajo:
        return "baja"
    elif bajo < datos <= alto:
        return "moderada"
    else:
        return "alta"
```

```python
df['infant_mortality_level'] = df['infant_mortality'].apply(mortalidad_inf)
df['infant_mortality_level']
df
```
